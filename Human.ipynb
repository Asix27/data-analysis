```python
#Предобработка
from sklearn.preprocessing import LabelEncoder

# Создание объекта LabelEncoder
label_encoder = LabelEncoder()

# Применение Label Encoding к колонке "RaceDesc"
RaceDesc_Encoded = label_encoder.fit_transform(df['RaceDesc'])
race_desc_series = pd.Series(RaceDesc_Encoded, name='RaceDesc_Encoded')

features_name = [
    'MarriedID', 'MaritalStatusID', 'EmpStatusID', 'DeptID', 'PerfScoreID', 
    'FromDiversityJobFairID', 'Salary', 'Termd', 'PositionID', 
    'ManagerID', 
    'EngagementSurvey', 'EmpSatisfaction', 'SpecialProjectsCount', 
    'DaysLateLast30', 'Absences'
]

X = pd.concat([df[features_name],race_desc_series], axis=1)

X.shape
X
```

```python
# Расчет среднего уровня вовлеченности для каждого ManagerID
mean_engagement = df.groupby('ManagerID')['EngagementSurvey'].mean().reset_index()

# Визуализация среднего уровня вовлеченности
plt.figure(figsize=(10,7))
sns.barplot(x='ManagerID', y='EngagementSurvey', data=mean_engagement)
plt.title('Средний уровень вовлеченности по менеджерам')
plt.show()

# Для расчета медианы просто замените .mean() на .median()
median_engagement = df.groupby('ManagerID')['EngagementSurvey'].median().reset_index()

# Визуализация медианного уровня вовлеченности
plt.figure(figsize=(10,7))
sns.barplot(x='ManagerID', y='EngagementSurvey', data=median_engagement)
plt.title('Медианный уровень вовлеченности по менеджерам')
plt.show()
```

```python
# Расчет среднего уровня опоздание
mean_engagement = X.groupby('RaceDesc_Encoded')['Absences'].mean().reset_index()

# Визуализация среднего уровня опоздание
plt.figure(figsize=(10,7))
sns.barplot(x='RaceDesc_Encoded', y='Absences', data=mean_engagement)
plt.title('Средний уровень вовлеченности по менеджерам')
plt.show()

# Для расчета медианы просто замените .mean() на .median()
median_engagement = X.groupby('RaceDesc_Encoded')['Absences'].median().reset_index()

# Визуализация медианного уровня опоздание
plt.figure(figsize=(10,7))
sns.barplot(x='RaceDesc_Encoded', y='Absences', data=median_engagement)
plt.title('Медианный уровень вовлеченности по менеджерам')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
import numpy as np

# Выбор данных для анализа
X = df[['Salary']]  # Предиктор (независимая переменная)
y = df['MarriedID']  # Целевая переменная (зависимая переменная)

# Построение модели логистической регрессии
log_reg = LogisticRegression()
log_reg.fit(X, y)

# Создание последовательности значений зарплат для предсказаний
X_test = np.linspace(X.min(), X.max(), 300)

# Предсказание вероятностей для каждого значения в X_test
y_prob = log_reg.predict_proba(X_test)[:,1]

# Визуализация
plt.figure(figsize=(10, 6))
plt.scatter(X, y, alpha=0.5)
plt.plot(X_test, y_prob, color='red')  # Линия логистической регрессии
plt.title('Логистическая регрессия: Зарплата vs Семейное положение')
plt.xlabel('Зарплата')
plt.ylabel('Вероятность замужества/женитьбы')
plt.show()
```
